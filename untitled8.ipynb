```python
townName = []
matchesplayed = []
TownData = df['home_team'].value_counts()
for _ in TownData.index:
    if TownData[_] > 30:
        townName.append(_)
        matchesplayed.append(TownData[_])
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(townName,matchesplayed)
plt.xlabel('Town Name')
plt.ylabel('Matches Played')
plt.title('Matches Played by Town')
plt.show()
```

```python
def matchesplayedbyTown():
  townName = []
  matchesplayed = []
  TownData = df['home_team'].value_counts()
  for _ in TownData.index:
    if TownData[_] > 30:
        townName.append(_)
        matchesplayed.append(TownData[_])
    plt.barh(townName,matchesplayed)
    plt.xlabel('Town Name')
    plt.ylabel('Matches Played')
    plt.title('Matches Played by Town')
    plt.show()
```

```python
matchesplayedbyTown()
```

```python
plt.bar(townName,matchesplayed)
plt.xlabel('Town Name')
plt.ylabel('Matches Played')
plt.title('Matches Played by Town')
plt.show()
```
